priors_table.tex

```python
dfs = []
for priors in [('RBC', mod4_priors), ('NK',  mod5_priors), ('NK Energy', mod6_priors)]:
    df = pd.DataFrame(
        [
            chain(
                [f'{item}', priors[1][item].dist.name], [*priors[1][item].stats()]
            ) for item in priors[1].keys()
        ],
        columns=['param', 'distribution', 'mean', 'std']
    )

    df['model'] = list([priors[0]] * len(df))
    df = df[['model', 'param', 'distribution', 'mean', 'std']]
    
    dfs.append(df)

dfs = pd.concat(dfs)
dfs = dfs.pivot_table(columns=['model'], index=['param'], values=['mean', 'std'])
dfs.fillna('', inplace=True)
```

```python
with open(os.path.join(LATEX_OUT, 'priors_table.tex'), 'w') as tf:
    tf.write(dfs.to_latex(caption='Prior distributions'))
```

```python
from io import StringIO 
import sys

class Capturing(list):
    def __enter__(self):
        self._stdout = sys.stdout
        sys.stdout = self._stringio = StringIO()
        return self
    def __exit__(self, *args):
        self.extend(self._stringio.getvalue().splitlines())
        del self._stringio    # free up some memory
        sys.stdout = self._stdout
```

```python
with Capturing() as output:
    mod.print_steady_state()
```

```python
df = pd.DataFrame([item.split('           ') for item in output]).drop(6, axis=0)

with open(os.path.join(LATEX_OUT, 'rbc_steady_state.tex'), 'w') as tf:
    tf.write(df.to_latex(caption='RBC steady state values'))
df
```
